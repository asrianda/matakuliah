```python
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import MinMaxScaler

# Dataset
X = np.array([
    [2, 60, 50],
    [3, 65, 55],
    [4, 70, 60],
    [5, 75, 65],
    [6, 80, 70],
    [7, 85, 75],
    [8, 90, 80],
    [1, 50, 45],
    [9, 95, 85],
    [10, 98, 90]
])

y = np.array([0,0,0,1,1,1,1,0,1,1])

# Normalisasi (lebih standar pakai scaler)
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# Model Neural Network
model = MLPClassifier(
    hidden_layer_sizes=(4,),   # 1 hidden layer, 4 neuron
    activation='relu',
    solver='adam',
    max_iter=1000,
    random_state=42
)

# Training
model.fit(X, y)

# Prediksi
pred = model.predict(X)

# Probabilitas
prob = model.predict_proba(X)

print("Probabilitas:")
print(prob)

print("\nPrediksi kelas:")
print(pred)
```

## **Neural Network**

```python
import numpy as np

# Dataset
X = np.array([
    [2, 60, 50],
    [3, 65, 55],
    [4, 70, 60],
    [5, 75, 65],
    [6, 80, 70],
    [7, 85, 75],
    [8, 90, 80],
    [1, 50, 45],
    [9, 95, 85],
    [10, 98, 90]
])

y = np.array([[0],[0],[0],[1],[1],[1],[1],[0],[1],[1]])

# Normalisasi
X = X / np.max(X, axis=0)

# Inisialisasi bobot
np.random.seed(1)
weights_input_hidden = np.random.rand(3, 2)
weights_hidden_output = np.random.rand(2, 1)

bias_hidden = np.random.rand(1, 2)
bias_output = np.random.rand(1, 1)

# Fungsi aktivasi
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Training
epochs = 5000
lr = 0.1

for i in range(epochs):

    # FORWARD
    hidden = sigmoid(np.dot(X, weights_input_hidden) + bias_hidden)
    output = sigmoid(np.dot(hidden, weights_hidden_output) + bias_output)

    # ERROR
    error = y - output

    # BACKPROPAGATION
    d_output = error * sigmoid_derivative(output)
    d_hidden = d_output.dot(weights_hidden_output.T) * sigmoid_derivative(hidden)

    # UPDATE
    weights_hidden_output += hidden.T.dot(d_output) * lr
    weights_input_hidden += X.T.dot(d_hidden) * lr

    bias_output += np.sum(d_output, axis=0) * lr
    bias_hidden += np.sum(d_hidden, axis=0) * lr

# HASIL
print("Output prediksi:")
print(output)

print("\nPrediksi kelas:")
print((output > 0.5).astype(int))
```

## **Deep Learning**

```python
import numpy as np

# =========================
# 1. DATA INPUT
# =========================

# Normalisasi data
x1 = 0.2
x2 = 0.5

# Target
target = 0

# =========================
# 2. INISIALISASI BOBOT
# =========================

# Input -> Hidden
w1 = 0.1
w2 = 0.2
w3 = 0.3
w4 = 0.4

# Bias hidden
b1 = 0.1
b2 = 0.1

# Hidden -> Output
w5 = 0.5
w6 = 0.6

# Bias output
b3 = 0.1

# Learning rate
alpha = 0.1

# =========================
# 3. FUNGSI SIGMOID
# =========================

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# =========================
# 4. FORWARD PROPAGATION
# =========================

print("=== FORWARD PROPAGATION ===")

# Hidden neuron 1
zh1 = (w1 * x1) + (w2 * x2) + b1
h1 = sigmoid(zh1)

print("\nHidden Neuron 1")
print("zh1 =", round(zh1, 4))
print("h1 =", round(h1, 4))

# Hidden neuron 2
zh2 = (w3 * x1) + (w4 * x2) + b2
h2 = sigmoid(zh2)

print("\nHidden Neuron 2")
print("zh2 =", round(zh2, 4))
print("h2 =", round(h2, 4))

# Output neuron
zo = (w5 * h1) + (w6 * h2) + b3
y_pred = sigmoid(zo)

print("\nOutput Layer")
print("zo =", round(zo, 4))
print("y_pred =", round(y_pred, 4))

# =========================
# 5. HITUNG ERROR
# =========================

print("\n=== ERROR ===")

error = 0.5 * ((target - y_pred) ** 2)

print("Error =", round(error, 4))

# =========================
# 6. BACKPROPAGATION
# =========================

print("\n=== BACKPROPAGATION ===")

# Delta output
delta_o = (target - y_pred) * y_pred * (1 - y_pred)

print("\nDelta Output")
print("delta_o =", round(delta_o, 4))

# =========================
# 7. UPDATE HIDDEN -> OUTPUT
# =========================

print("\nUpdate Bobot Hidden -> Output")

w5_new = w5 + (alpha * delta_o * h1)
w6_new = w6 + (alpha * delta_o * h2)

print("w5 lama =", w5)
print("w5 baru =", round(w5_new, 4))

print("w6 lama =", w6)
print("w6 baru =", round(w6_new, 4))

# =========================
# 8. DELTA HIDDEN
# =========================

delta_h1 = h1 * (1 - h1) * (delta_o * w5)
delta_h2 = h2 * (1 - h2) * (delta_o * w6)

print("\nDelta Hidden")
print("delta_h1 =", round(delta_h1, 4))
print("delta_h2 =", round(delta_h2, 4))

# =========================
# 9. UPDATE INPUT -> HIDDEN
# =========================

print("\nUpdate Bobot Input -> Hidden")

w1_new = w1 + (alpha * delta_h1 * x1)
w2_new = w2 + (alpha * delta_h1 * x2)

w3_new = w3 + (alpha * delta_h2 * x1)
w4_new = w4 + (alpha * delta_h2 * x2)

print("\nw1 lama =", w1)
print("w1 baru =", round(w1_new, 4))

print("\nw2 lama =", w2)
print("w2 baru =", round(w2_new, 4))

print("\nw3 lama =", w3)
print("w3 baru =", round(w3_new, 4))

print("\nw4 lama =", w4)
print("w4 baru =", round(w4_new, 4))

# =========================
# 10. HASIL AKHIR
# =========================

print("\n=== HASIL AKHIR ITERASI ===")

print(f"w1 : {w1} -> {round(w1_new,4)}")
print(f"w2 : {w2} -> {round(w2_new,4)}")
print(f"w3 : {w3} -> {round(w3_new,4)}")
print(f"w4 : {w4} -> {round(w4_new,4)}")
print(f"w5 : {w5} -> {round(w5_new,4)}")
print(f"w6 : {w6} -> {round(w6_new,4)}")
```
