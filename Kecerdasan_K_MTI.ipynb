```python
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import MinMaxScaler

# Dataset
X = np.array([
    [2, 60, 50],
    [3, 65, 55],
    [4, 70, 60],
    [5, 75, 65],
    [6, 80, 70],
    [7, 85, 75],
    [8, 90, 80],
    [1, 50, 45],
    [9, 95, 85],
    [10, 98, 90]
])

y = np.array([0,0,0,1,1,1,1,0,1,1])

# Normalisasi (lebih standar pakai scaler)
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

# Model Neural Network
model = MLPClassifier(
    hidden_layer_sizes=(4,),   # 1 hidden layer, 4 neuron
    activation='relu',
    solver='adam',
    max_iter=1000,
    random_state=42
)

# Training
model.fit(X, y)

# Prediksi
pred = model.predict(X)

# Probabilitas
prob = model.predict_proba(X)

print("Probabilitas:")
print(prob)

print("\nPrediksi kelas:")
print(pred)
```

## **Neural Network**

```python
import numpy as np

# Dataset
X = np.array([
    [2, 60, 50],
    [3, 65, 55],
    [4, 70, 60],
    [5, 75, 65],
    [6, 80, 70],
    [7, 85, 75],
    [8, 90, 80],
    [1, 50, 45],
    [9, 95, 85],
    [10, 98, 90]
])

y = np.array([[0],[0],[0],[1],[1],[1],[1],[0],[1],[1]])

# Normalisasi
X = X / np.max(X, axis=0)

# Inisialisasi bobot
np.random.seed(1)
weights_input_hidden = np.random.rand(3, 2)
weights_hidden_output = np.random.rand(2, 1)

bias_hidden = np.random.rand(1, 2)
bias_output = np.random.rand(1, 1)

# Fungsi aktivasi
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Training
epochs = 5000
lr = 0.1

for i in range(epochs):

    # FORWARD
    hidden = sigmoid(np.dot(X, weights_input_hidden) + bias_hidden)
    output = sigmoid(np.dot(hidden, weights_hidden_output) + bias_output)

    # ERROR
    error = y - output

    # BACKPROPAGATION
    d_output = error * sigmoid_derivative(output)
    d_hidden = d_output.dot(weights_hidden_output.T) * sigmoid_derivative(hidden)

    # UPDATE
    weights_hidden_output += hidden.T.dot(d_output) * lr
    weights_input_hidden += X.T.dot(d_hidden) * lr

    bias_output += np.sum(d_output, axis=0) * lr
    bias_hidden += np.sum(d_hidden, axis=0) * lr

# HASIL
print("Output prediksi:")
print(output)

print("\nPrediksi kelas:")
print((output > 0.5).astype(int))
```
